```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
pd.__version__
```

```python
mappings = {
'India':1,
'Australia':2,
'England':3,
'South Africa':4,
'New Zealand':5,
'Pakistan':6,
'West Indies':7,
'Sri Lanka':8,
'Bangladesh':9,
'Afganistan':10
}
```

```python
data = pd.read_excel (r'C:\Users\Dhruv\Desktop\WC.xlsx')
```

```python
data
```

```python
df = pd.DataFrame(data)
df.index = np.arange(1, len(df)+1)
array_of_cols = df.columns.values
for x in array_of_cols:
    df[x].replace(mappings, inplace=True)
df
```

```python
df_X = df.loc[:,'Team 1':'Team 4']
df_Y = df.loc[:,'Winner']
X = df_X.to_numpy()
Y = df_Y.to_numpy()
```

```python
X = X.astype('int')
X
```

```python
Y = Y.astype('int')
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
clf.fit(X_test, y_test)
```

```python
acc_train = clf.score(X_train, y_train)
acc_test = clf.score(X_test, y_test)
```

```python
print("Training set accuracy: {:.2f}".format(acc_train))  
print("Test set accuracy: {:.2f}".format(acc_test))
```

```python
print("New prediction:", list(mappings.keys())[list(mappings.values()).index(clf.predict([[mappings['England'],mappings['India'],mappings['Australia'],mappings['South Africa']]]))])
```
